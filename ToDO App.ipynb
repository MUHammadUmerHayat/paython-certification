```python
# ToDO App using List
toDoList = [];

while True:
    print("Task in our list:");
    for task in toDoList:
        print("* "+task);
        
    print("\n Options");
    print("1. Add task");
    print("2. Delete task");
    print("3. Delete all tasks");
    print("4. Quit");
    action = input("Enter your choice: ");
    
    if action == '1':
        toDoList.append(input("Enter your task: "));
    elif action == '2':
        toDoList.remove(input("Enter task to delete: "));
    elif action == '3':
        toDoList.clear();
    elif action == '4':
        break;
        
    print("\nxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx\n");
```

```python
# ToDO App using dictionary 
toDoList = {};
taskIndex = 0;
menu = {1:"Add task", 2:"Edit task", 3:"Delete task", 4:"Delete all tasks", 5:"Quit App"};


def dispalyMenu():
    print("\n--------- Main Menu ---------\n");
    for k,v in menu.items():
        print("{}. {}".format(k,v));

        
def isInt(userInput):
    try:
        int(userInput);
        return True;
    except ValueError:
        return False;
    
def getUserChoice():
    while(True):
        dispalyMenu();
        action = input("Enter your command: ");
        if isInt(action) == True:
            action = int(action);
            if(action > 0 and action <= 5):
                return action;
            
        print("\nInvalid Command, Try valid command from given options.");
            
def displayTasks():
    print("\n--------- Tasks list ---------\n");
    if len(toDoList) <= 0:
        print("No task in list");
        return True;
    for index, task in toDoList.items():
        print("Task #{} - {} ".format(index, task));
        
def addTask():
    global taskIndex;
    taskIndex += 1;
    task = input("Enter your task: ");
    toDoList.update({taskIndex:task});
    print("Task has added.");
    
def editTask():
    index = int(input("Enter task index to edit: "));
    if index in toDoList.keys():
        task = input("Enter new task: ");
        toDoList.update({index:task});
        print("Task has been updated.");
    else:
        print("Task not found.");
        
def deleteTask():
    index = int(input("Enter task index to delete: "));
    if index in toDoList.keys():
        toDoList.pop(index);
        print("Task has been deleted.");
    else:
        print("Task not found.");

while(True):
    displayTasks();
    action = getUserChoice();

    if action == 1:
        addTask();
    elif action == 2:
        editTask();
    elif action == 3:
        deleteTask();
    elif action == 4:
        toDoList.clear();
        print("All tasks are removed.");
    elif action == 5:
        break;
        
    print("\nxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx");
    print("xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx");
    print("xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx");
```
