```python
testCases = {};
def is_prime(n): # traditional method of finding prime number
    if (n <= 1):
        return False;
    elif (n==2):
        return True;
    else:
        counter = 0;
        for x in range(2, n):
            counter += 1;
            if(n % x == 0):
                return False;
        
        testCases[n]['p1c'] =  counter;
        return True;  

    
def is_prime2(n): # level 1 optimised method we skip even numbers
    if (n <= 1):
        return False;
    elif (n==2):
        return True;
    else:
        counter = 1;
        if(n % 2 == 0):
            return False;
        for x in range(3, n, 2):
            counter += 1;
            if(n % x == 0):
                return False;
        
        testCases[n]['p2c'] =  counter;
        return True;  
    

for x in range(1, 10000):
    testCases[x] = {};
    pRes = is_prime(x);
    p2Res = is_prime2(x);
    if(len(testCases[x]) == 0):
        del(testCases[x]);
    else:
        print(x, testCases[x]);
    #if(pRes['res'] or p2Res['res']):
        #testCases[x] = {'p1':pRes, 'p2':p2Res};

```
